#### 3) คะแนนเฉลี่ยของ HomeTeam และ AwayTeam มีความแตกต่างกันเท่าไหร่

```python
average_home_goals = df['FTHG'].mean()
average_away_goals = df['FTAG'].mean()

goal_difference = abs(average_home_goals - average_away_goals)

print(f"Average number of HomeTeam: {average_home_goals}")
print(f"Average number of AwayTeam: {average_away_goals}")
print(f"Difference: {goal_difference}")
```

### Advanced 2 question

### สัดส่วนการชนะของ HomeTeam และ AwayTeam

```python
df['Result'] = 'Draw'  # สมมติว่าเริ่มต้นทั้งหมดเป็นการเสมอ
df.loc[df['FTR'] == 'H', 'Result'] = 'Home Win'  # ทีมเจ้าบ้านชนะ
df.loc[df['FTR'] == 'A', 'Result'] = 'Away Win'  # ทีมเยือนชนะ

# นับจำนวนการชนะ
home_wins = df[df['Result'] == 'Home Win'].shape[0]
away_wins = df[df['Result'] == 'Away Win'].shape[0]

# คำนวณสัดส่วนการชนะ
total_matches = df.shape[0]
home_win_percentage = (home_wins / total_matches) * 100
away_win_percentage = (away_wins / total_matches) * 100


print(f'Home Win: {home_win_percentage:.2f}%')
print(f'Away Win: {away_win_percentage:.2f}%')
```

### วิเคราะห์ความสัมพันธ์ระหว่าง FTHG และ HST ว่าความสัมพันธ์นี้เป็นเชิงบวกหรือลบ?

```python
import numpy as np
correlation_coefficient = np.corrcoef(df['FTHG'], df['HST'])[0, 1]
print(f"Correlation Coefficient between FTHG and HST: {correlation_coefficient}")
```
